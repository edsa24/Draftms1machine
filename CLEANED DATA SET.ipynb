```python
import pandas as pd

demographics = pd.read_csv('customer_demographics_contaminated.csv')
transactions = pd.read_csv('customer_transactions_contaminated.csv')
social = pd.read_csv('social_media_interactions_contaminated.csv')

print("Demographics:")
display(demographics.head())

print("\nTransactions:")
display(transactions.head())

print("\nSocial Media Interactions:")
display(social.head())
```

```python
# Check general information: column names, data types, and non-null counts
print("=== Demographics Info ===")
demographics.info()

print("\n=== Transactions Info ===")
transactions.info()

print("\n=== Social Media Info ===")
social.info()

# View sample entries
print("\n=== Sample Entries (Demographics) ===")
display(demographics.sample(5))

print("\n=== Sample Entries (Transactions) ===")
display(transactions.sample(5))

print("\n=== Sample Entries (Social Media) ===")
display(social.sample(5))
```

```python
# Check duplicates count in each dataset
print("Duplicates in Demographics:", demographics.duplicated().sum())
print("Duplicates in Transactions:", transactions.duplicated().sum())
print("Duplicates in Social Media:", social.duplicated().sum())

# Remove duplicates
demographics = demographics.drop_duplicates()
transactions = transactions.drop_duplicates()
social = social.drop_duplicates()

print("\nAfter removing duplicates:")
print("Demographics shape:", demographics.shape)
print("Transactions shape:", transactions.shape)
print("Social Media shape:", social.shape)
```

```python
# Check missing values for each dataset
print("=== Missing Values in Demographics ===")
print(demographics.isnull().sum())

print("\n=== Missing Values in Transactions ===")
print(transactions.isnull().sum())

print("\n=== Missing Values in Social Media ===")
print(social.isnull().sum())

# Optional: show percentage of missing values
print("\n=== Missing Value Percentage ===")
for name, df in zip(
    ['Demographics', 'Transactions', 'Social Media'],
    [demographics, transactions, social]
):
    print(f"\n{name}:")
    print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
# Save cleaned datasets
demographics.to_csv('customer_demographics_clean.csv', index=False)
transactions.to_csv('customer_transactions_clean.csv', index=False)
social.to_csv('social_media_interactions_clean.csv', index=False)

print("Cleaned datasets have been saved successfully!")
```
